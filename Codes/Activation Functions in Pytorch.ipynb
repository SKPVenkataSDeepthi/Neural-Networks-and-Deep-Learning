```python
## ReLU Activation Function
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Defining ReLU
r = nn.ReLU()

# Creating a range of values
input = torch.linspace(-10, 10, 100)

# Passing the array to ReLU function
output = r(input)

# Plotting the input and output
input_array = input.numpy()
output_array = output.detach().numpy()

plt.figure()
plt.plot(input_array, output_array, label='ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Leaky ReLU function
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Defining Leaky ReLU with a negative slope of 0.2
r = nn.LeakyReLU(0.2)

# Creating a range of values
input = torch.linspace(-10, 10, 100)

# Passing the array to Leaky ReLU function
output = r(input)

# Plotting the input and output
input_array = input.numpy()
output_array = output.detach().numpy()

plt.figure()
plt.plot(input_array, output_array, label='Leaky ReLU Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Leaky ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Sigmoid Function
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Calling the sigmoid function
sig = nn.Sigmoid()

# Defining tensor
input = torch.linspace(-10, 10, 100)

# Applying sigmoid to the tensor
output = sig(input)

print(output)

# Plotting the input and output
input_array = input.numpy()
output_array = output.detach().numpy()

plt.figure()
plt.plot(input_array, output_array, label='Sigmoid Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Tanh Function
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Calling the Tanh function
t = nn.Tanh()

# Defining tensor
input = torch.linspace(-10, 10, 100)

# Applying Tanh to the tensor
output = t(input)

# Plotting the input and output
input_array = input.numpy()
output_array = output.detach().numpy()

plt.figure()
plt.plot(input_array, output_array, label='Tanh Activation')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Tanh Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Sigmoid vs tanh
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Defining Sigmoid and Tanh functions
sigmoid = nn.Sigmoid()
tanh = nn.Tanh()

# Creating a range of values
input = torch.linspace(-10, 10, 100)

# Applying Sigmoid and Tanh to the tensor
output_sigmoid = sigmoid(input)
output_tanh = tanh(input)

# Plotting the input and output for both functions
input_array = input.numpy()
output_sigmoid_array = output_sigmoid.detach().numpy()
output_tanh_array = output_tanh.detach().numpy()

plt.figure()
plt.plot(input_array, output_sigmoid_array, label='Sigmoid Activation')
plt.plot(input_array, output_tanh_array, label='Tanh Activation', linestyle='--')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid vs Tanh Activation Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Softmax function
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np

# Calling the Softmax function with dimension = 0
sm = nn.Softmax(dim=0)  

# Defining tensor
input = torch.Tensor([1, -2, 3, -5]) 

# Applying function to the tensor
output = sm(input)

# Plotting the input and output
input_array = input.numpy()
output_array = output.detach().numpy()

plt.figure()
plt.bar(range(len(input_array)), output_array, label='Softmax Activation')
plt.xlabel('Input Index')
plt.ylabel('Probability')
plt.title('Softmax Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```
